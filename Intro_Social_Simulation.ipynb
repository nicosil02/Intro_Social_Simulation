```python
import pandas as pd
strategies=['Rock','Paper','Scissors']
```

## Rules

The rules tell you that according to a strategy followed, players get a pay-off:

```python
payoff={('Rock','Paper'):(0,1), #instituciones
        ('Paper','Rock'):(1,0),
        ('Rock','Scissors'):(1,0),
        ('Scissors','Rock'):(0,1),
        ('Paper','Scissors'):(0,1),
        ('Scissors','Paper'):(1,0),
        ('Rock','Rock'):(0,0),
        ('Paper','Paper'):(0,0),
        ('Scissors','Scissors'):(0,0)}
```

## Creating and setting up agents:

Players have a name, but have no score, and no strategy yet.

## Decision making process

This is the process to choose an strategy:

```python
from random import choice

#simplest strategy: choose randomly
choice(strategies);
```

```python
# names of players
names=['Jim','Jane','Peter','Zoe']
```

```python
# setting up players
society=[{'name':n,'score':0,'strategy':None} for n in names]
```

```python
# each player a dict:
society
```

```python
import itertools

# pair is a tuple of dicts
for pair in itertools.combinations(society,2):
    print(pair)
```

```python
import itertools

# each dict
for player1,player2 in itertools.combinations(society,2):
    print(player1,player2)
```

```python
society = [{'name': n, 'score': 0, 'strategy': None} for n in names]

for aRound in range(100):
    print(f"--- Round {aRound + 1} ---")

    for player1, player2 in itertools.combinations(society, 2):
        player1['strategy'] = choice(strategies)
        player2['strategy'] = choice(strategies)

        result = payoff[player1['strategy'], player2['strategy']]

        player1['score'] += result[0]
        player2['score'] += result[1]

        # determine the winner of the current match
        if result[0] > result[1]:
            winner = player1['name']
        elif result[1] > result[0]:
            winner = player2['name']
        else:
            winner = "Tie"

        print(f"{player1['name']} chose {player1['strategy']}, {player2['name']} chose {player2['strategy']}. The winner is: {winner}. Current scores: {player1['name']} {player1['score']} - {player2['name']} {player2['score']}")

print("\n--- Final Scores ---")
for player in society:
    print(f"{player['name']}: {player['score']}")

max_score = max(player['score'] for player in society)
winners = [player['name'] for player in society if player['score'] == max_score]

print("\n--- Final Winner(s) ---")
if len(winners) > 1:
    print(f"It's a tie! The winners are: {', '.join(winners)} with {max_score} points each.")
else:
    print(f"The overall winner is: {winners[0]} with {max_score} points.")
```

```python
# as a data frame
socialResults=pd.DataFrame(society)
socialResults
```

```python
# names of players
names=['Jim','Jane','Peter','Zoe',"Juanito","Alberto","Nicolas","Pedro","Enrique","Rosite"]
```

```python
strategies = ['Rock', 'Paper', 'Scissors', 'Lizard', 'Spock']
```

```python
payoff = {
    ('Rock', 'Scissors'): (1, 0),
    ('Rock', 'Lizard'): (1, 0),
    ('Scissors', 'Paper'): (1, 0),
    ('Scissors', 'Lizard'): (1, 0),
    ('Paper', 'Rock'): (1, 0),
    ('Paper', 'Spock'): (1, 0),
    ('Lizard', 'Spock'): (1, 0),
    ('Lizard', 'Paper'): (1, 0),
    ('Spock', 'Scissors'): (1, 0),
    ('Spock', 'Rock'): (1, 0),
    ('Scissors', 'Rock'): (0, 1),
    ('Lizard', 'Rock'): (0, 1),
    ('Paper', 'Scissors'): (0, 1),
    ('Lizard', 'Scissors'): (0, 1),
    ('Rock', 'Paper'): (0, 1),
    ('Spock', 'Paper'): (0, 1),
    ('Spock', 'Lizard'): (0, 1),
    ('Paper', 'Lizard'): (0, 1),
    ('Scissors', 'Spock'): (0, 1),
    ('Rock', 'Spock'): (0, 1),
    ('Rock', 'Rock'): (0, 0),
    ('Paper', 'Paper'): (0, 0),
    ('Scissors', 'Scissors'): (0, 0),
    ('Lizard', 'Lizard'): (0, 0),
    ('Spock', 'Spock'): (0, 0),
}
```

```python
society=[{'name':n,'score':0,'strategy':None} for n in names]
```

```python
import itertools

# pair is a tuple of dicts
for pair in itertools.combinations(society,2):
    print(pair)
```

```python
# each dict
for player1,player2 in itertools.combinations(society,2):
    print(player1,player2)
```

```python
# Sheldon Tournament
for aRound in range(100):
    print(f"--- Round {aRound + 1} ---")

    for player1, player2 in itertools.combinations(society, 2):
        player1['strategy'] = choice(strategies)
        player2['strategy'] = choice(strategies)

        result = payoff.get((player1['strategy'], player2['strategy']), (0, 0))

        player1['score'] += result[0]
        player2['score'] += result[1]

        # Determine winner
        if result[0] > result[1]:
            winner = player1['name']
        elif result[1] > result[0]:
            winner = player2['name']
        else:
            winner = "Tie"

        print(f"{player1['name']} chose {player1['strategy']}, {player2['name']} chose {player2['strategy']}. The winner is: {winner}. Current scores: {player1['name']} {player1['score']} - {player2['name']} {player2['score']}")

    print()

print("\n--- Final Scores ---")
for player in society:
    print(f"{player['name']}: {player['score']}")

max_score = max(player['score'] for player in society)
winners = [player['name'] for player in society if player['score'] == max_score]

print("\n--- Final Winner(s) ---")
if len(winners) > 1:
    print(f"It's a tie! The winners are: {', '.join(winners)} with {max_score} points each.")
else:
    print(f"The overall winner is: {winners[0]} with {max_score} points.")
```
